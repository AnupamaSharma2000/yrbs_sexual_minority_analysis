Three types of missing data:
1. Missing completely at random, MCAR

        When the missingness is unrelated to any variable or the value itself — typically caused by data-entry or recording errors, in this case participant might just be lazy to fill the survey.
2. Missing at random, MAR

       When missingness depends on other observed variables, not on the value itself

3. Missing not at random, MNAR

       When missingness depends on the value itself — common with sensitive or stigmatized topics.

#Transformations -- New Variables by combining

````python
#Creating Sexual minority using sexual contact variable

def assign_sexcontact(row):
    sex = row['Sex']
    q63 = row['q63']

    if np.isnan(sex) or np.isnan(q63):
        return None
    sex = int(sex)
    q63 = int(q63)

    if sex == 1:  # Female
        if q63 == 1: return 4  # No sexual contact
        elif q63 == 2: return 2  # Female contact (same sex)
        elif q63 == 3: return 1  # Male contact (opposite sex)
        elif q63 == 4: return 3  # Both
    elif sex == 2:  # Male
        if q63 == 1: return 4
        elif q63 == 2: return 1
        elif q63 == 3: return 2
        elif q63 == 4: return 3
    return None

    # Changin dtype to integers if not NaN
raw_data['sexcontact'] = raw_data.apply(assign_sexcontact, axis=1)

sexcontact_labels = {
    1: 'Opposite sex only',
    2: 'Same sex only',
    3: 'Both sexes',
    4: 'No sexual contact'
}
sex_labels = {
    1: 'Female',
    2: 'Male'
}

raw_data['sexcontact_label'] = raw_data['sexcontact'].map(sexcontact_labels)
raw_data['sex_label'] = raw_data['Sex'].map(sex_labels)

# Dropping NaNs for analysis and plot
plot_data = raw_data.dropna(subset=['sexcontact_label', 'sex_label'])

fig = px.histogram(
    plot_data,
    y='sexcontact_label',
    color='sex_label',
    text_auto=True,
    barnorm=None,
    category_orders={"sexcontact_label": ['No sexual contact', 'Opposite sex only', 'Same sex only', 'Both sexes']},
    labels={"sexcontact_label": "Sexual Contact Category", "count": "Count", "sex_label": "Sex"},
    title="Distribution of Sexual Contact by Sex"
)

fig.update_traces(
    hovertemplate='%{y}<br>Count: %{x}'
)
fig.show()



````

Combining Tobacco_use variable

````python
def tobacco_intensity(row):

    #all products not used, or never experiment (q31, q35==2) and all day-use == 1 implies NO USE
    no_use = (
        (row['q31'] == 2 or pd.isna(row['q31'])) and
        (row['q35'] == 2 or pd.isna(row['q35'])) and
        all((row[v] == 1 or pd.isna(row[v])) for v in ['q33', 'q36', 'q38', 'q39'])
    )
    if no_use:
        return 'No use'


    #used any product >=6 days in past 30, or vaped/smoked/chewed a lot
    heavy_use = any(
        (row[v] is not np.nan) and (row[v] >= 4)
        for v in ['q33', 'q36', 'q38', 'q39']
    )
    if heavy_use:
        return 'Frequent/Heavy use'


    #if not no_use or heavy_use, must be light/experimental (any 2-5 days use or ever tried)
    lite_ever = (
        (row['q31'] == 1 or row['q35'] == 1) or
        any((row[v] in [2, 3])
            for v in ['q33', 'q36', 'q38', 'q39']
            if row[v] is not np.nan)
    )
    if lite_ever:
        return 'Light/Experimental use'

    #if missing completely or unmatched, marked as missing
    return 'Missing'


raw_data['tobacco_use_intensity'] = raw_data.apply(tobacco_intensity, axis=1)


print(raw_data['tobacco_use_intensity'].value_counts(dropna=False))
````

````python

def alcohol_intensity(row):

    #never drank and no alcohol use or binge in past 30 days
    no_use = (
        (row['q41'] == 1 or pd.isna(row['q41'])) and
        (row['q42'] == 1 or pd.isna(row['q42'])) and
        (row['q43'] == 1 or pd.isna(row['q43'])) and
        (row['q44'] == 1 or pd.isna(row['q44']))
    )
    if no_use:
        return 'No use'

    #any high-frequency drinking or binge ("6+ days" codes, i.e. q42/q43 >=4), or high max drinks (q44 >=6)
    heavy = (
        (row['q42'] is not np.nan and row['q42'] >= 4) or
        (row['q43'] is not np.nan and row['q43'] >= 4) or
        (row['q44'] is not np.nan and row['q44'] >= 6)
    )
    if heavy:
        return 'Frequent/Heavy use'


    #has tried, and low-frequency use or binge last 30 ("few days": codes 2 or 3)
    light = (
        (row['q41'] != 1 and not pd.isna(row['q41'])) or
        (row['q42'] in [2, 3]) or
        (row['q43'] in [2, 3])
    )
    if light:
        return 'Light/Experimental use'


    return 'Missing'


raw_data['alcohol_use_intensity'] = raw_data.apply(alcohol_intensity, axis=1)

print(raw_data['alcohol_use_intensity'].value_counts(dropna=False))
````

Combining Marijuana, pain medicine, illicit drugs variable

````python
drug_vars = ['q46', 'q47', 'q48', 'q49', 'q50', 'q51', 'q52', 'q53', 'q54', 'q55', 'q92', 'q93']

def drug_intensity(row):

    no_use = all((row[v] == 1 or pd.isna(row[v])) for v in drug_vars)
    if no_use:
        return 'No use'

    #frequent/Heavy Use (any substance used 10+ times, i.e., code >=4)
    heavy_use = any((row[v] is not np.nan and row[v] >= 4) for v in drug_vars)
    if heavy_use:
        return 'Frequent/Heavy use'

    #light/Experimental Use (at least one code 2–3, and not heavy)
    light_exp = any((row[v] in [2,3]) for v in drug_vars if row[v] is not np.nan)
    if light_exp:
        return 'Light/Experimental use'

    #incomplete response
    return 'Missing'

raw_data['drug_use_intensity'] = raw_data.apply(drug_intensity, axis=1)


print(raw_data['drug_use_intensity'].value_counts(dropna=False))
````

````python
from scipy.stats import chi2_contingency
import itertools

def cramers_v(confusion_matrix):
    """Calculate Cramér's V statistic for categorical-categorical association."""
    chi2 = chi2_contingency(confusion_matrix, correction=False)[0]
    n = confusion_matrix.sum().sum()
    if n == 0: return np.nan
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k)-1)))

def variable_relationship_report(df, variables):

    print("\nEvent frequencies and value distributions:")
    for var in variables:
        print(f"\nVariable: {var}\n", df[var].value_counts(dropna=False))
        plt.figure()
        if df[var].dtype == 'object' or df[var].nunique() < 10:
            sns.countplot(y=df[var], order=df[var].value_counts().index)
        else:
            sns.histplot(df[var].dropna(), kde=False)
        plt.title(f"{var} - Distribution")
        plt.show()





    print("\nPairwise categorical associations (Cramér’s V):")
    pairs = list(itertools.combinations(variables, 2))
    results = []
    for v1, v2 in pairs:
        if df[v1].nunique() < 30 and df[v2].nunique() < 30:
            ctab = pd.crosstab(df[v1], df[v2])
            v = cramers_v(ctab)
            results.append({'var1': v1, 'var2': v2, 'cramers_v': v})
    cramer_df = pd.DataFrame(results).pivot(index='var1', columns='var2', values='cramers_v')
    print(cramer_df)
    plt.figure(figsize=(12,10))
    sns.heatmap(cramer_df, annot=True, cmap='YlGnBu')
    plt.title("Cramér's V Heatmap (Categorical Correlation)")
    plt.show()


    numeric_vars = [v for v in variables if np.issubdtype(df[v].dtype, np.number)]
    if len(numeric_vars) > 1:
        print("\nSpearman correlations (numeric variables):")
        corrmat = df[numeric_vars].corr(method="spearman")
        print(corrmat)
        plt.figure(figsize=(10,8))
        sns.heatmap(corrmat, annot=True, cmap="coolwarm")
        plt.title("Spearman Correlation (Numeric)")
        plt.show()


    print("\nSelected Pairwise Relationships/Crosstabs:")
    for v1, v2 in pairs[:5]:   # Only show a few for brevity
        if df[v1].nunique() < 5 and df[v2].nunique() < 5:
            ctab = pd.crosstab(df[v1], df[v2])
            print(f"\nCrosstab of {v1} x {v2}:\n", ctab)
            sns.heatmap(ctab, annot=True)
            plt.title(f"Crosstab: {v1} x {v2}")
            plt.show()

    print("\nInterpretation:\n- Pairs with Cramér’s V > 0.3 show strong dependency.\n- Most other pairs with low values are likely independent.\n")

variables = [
 'Age',
 'Sex',
 'Grade',
 'Height_meters',
 'Weight_kg',
 'sex_iden_label',
 'sexcontact_label',
 'q63',
 'raceeth_label',
 'sexcontact',
 'sex_label',
 'sexual_contact_minority_label',
 'sexual_contact_minority',
 'sex_iden_minority_label',
 'sex_iden_minority',
 'gender_minority',
 'sex_iden_gender_minority',
 'sex_iden_gender_minority_label',
 'bullying_experience',
 'bullying_experience_label',
 'mental_health_flag',
 'mental_health_score',
 'tobacco_use_intensity',
 'alcohol_use_intensity',
 'drug_use_intensity',

]
variable_relationship_report(dataset_1, variables)
````

````python
import plotly.express as px

def plot_by_subgroup(df, variable, subgroup_var, weight_col):

    overall = (
        df
        .dropna(subset=[variable, weight_col])
        .groupby(variable)[weight_col]
        .sum()
        .reset_index(name='weight_sum')
    )
    overall['percent'] = 100 * overall['weight_sum'] / overall['weight_sum'].sum()

    fig_overall = px.bar(
        overall,
        x=variable,
        y='percent',
        title=f'Overall weighted distribution of {variable}'
    )
    fig_overall.show()


    for subgroup in df[subgroup_var].dropna().unique():
        subgroup_data = df[df[subgroup_var] == subgroup]
        subgroup_dist = (
            subgroup_data
            .dropna(subset=[variable, weight_col])
            .groupby(variable)[weight_col]
            .sum()
            .reset_index(name='weight_sum')
        )
        subgroup_dist['percent'] = 100 * subgroup_dist['weight_sum'] / subgroup_dist['weight_sum'].sum()

        fig = px.bar(
            subgroup_dist,
            x=variable,
            y='percent',
            title=f'Weighted distribution of {variable} for {subgroup_var} = {subgroup}'
        )
        fig.show()


plot_by_subgroup(dataset_1, 'mental_health_flag', 'sex_iden_gender_minority_label', 'weight')


````

If subgroup trends contradict or reverse the overall trend, it indicates possible Simpson's Paradox.

Here we can that the overall trend is majority of participants reporting no mental health risk while in sexual and gender minority subgroup the trend reverses with majority reporting some mental health risk factor



````python
import plotly.express as px

def plot_weighted_stratified(df, var, subgroup_var, weight_var='weight'):

    overall = (
        df
        .dropna(subset=[var, weight_var])
        .groupby(var)[weight_var]
        .sum()
        .reset_index(name='weight_sum')
    )
    overall['percent'] = 100 * overall['weight_sum'] / overall['weight_sum'].sum()
    fig_overall = px.bar(
        overall,
        x=var,
        y='percent',
        title=f'Overall Weighted Distribution of {var}'
    )
    fig_overall.show()

    for group in df[subgroup_var].dropna().unique():
        subgroup_df = df[df[subgroup_var] == group]
        subgroup_dist = (
            subgroup_df
            .dropna(subset=[var, weight_var])
            .groupby(var)[weight_var]
            .sum()
            .reset_index(name='weight_sum')
        )
        subgroup_dist['percent'] = 100 * subgroup_dist['weight_sum'] / subgroup_dist['weight_sum'].sum()

        fig = px.bar(
            subgroup_dist,
            x=var,
            y='percent',
            title=f'Weighted Distribution of {var} within {subgroup_var} = {group}'
        )
        fig.show()


plot_weighted_stratified(raw_data, 'tobacco_use_intensity', 'Grade')


plot_weighted_stratified(raw_data, 'alcohol_use_intensity', 'Grade')
plot_weighted_stratified(raw_data, 'drug_use_intensity', 'Grade')
plot_weighted_stratified(raw_data, 'mental_health_flag', 'Grade')
````
