Three types of missing data:
1. Missing completely at random, MCAR

        When the missingness is unrelated to any variable or the value itself — typically caused by data-entry or recording errors, in this case participant might just be lazy to fill the survey.
2. Missing at random, MAR

       When missingness depends on other observed variables, not on the value itself

3. Missing not at random, MNAR

       When missingness depends on the value itself — common with sensitive or stigmatized topics.

Some missing data maybe because of Data edits done - see page 7 of guide
